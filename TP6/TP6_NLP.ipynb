## Asociación de palabras[1]


### 1. Levantar el corpus AP, separando cada noticia como un elemento distinto en un diccionario ( DOCNO :  TEXT).

```python
# LEVANTAR CORPUS AP
file_name = 'H:\\Cursos Doctorado\\CienciaDeDatos\\6.NLP\\ap\\ap.txt'

#load as string
with open(file_name, 'r') as content_file:
    content = content_file.read()

soup = BeautifulSoup(content, 'html.parser')

result_tokens = {} #dict with tokenized texts, WITHOUT stop words
result_complete = {} #dict with texts as strings (WITHOUT tokenization)
sib = soup.doc
tokenizer = RegexpTokenizer(r'\w+')
for d in range(0,len(soup.find_all('doc'))):
    key = str(sib.docno.string).strip()
    value = str(sib.contents[3].text)    #el atributo text tiene un problema porque hay una función con el mismo nombre    
    lowers = value.lower() #todas las palabras en lower case    
    result_complete[key] = lowers
    tokens = tokenizer.tokenize(lowers)   # es el mejor pero no maneja bien los contractions
    result_tokens[key] = [w for w in tokens if not w in stopwords.words('english')] #no stop words
    sib = sib.next_sibling.next_sibling

print(len(result_complete.keys())) 
```

### 2. Calcular el tamaño del vocabulario.


```python
#CALCULAR EL TAMAÑO DEL VOCABULARIO
all_tokens = [] #tokens without stop words

for key in result_complete:          
    all_tokens = all_tokens + result_tokens[key]

vocabulary_n = len(set(all_tokens))
print('Tamaño del vocabulario sin STOP words = ' + str(vocabulary_n))
```

### 3. Para las 500 palabras con más apariciones, calcular el par más asociado según la medida presentada.

```python
#Para las 500 palabras con más apariciones, calcular el par más asociado según la medida presentada
N = 500 #nro de palabras con más apariciones
win_size = 5 #tamaño de la venta
threshold = 6 #umbral sobre el cual se calcula pmi
#most frequent
f_distribution = nltk.probability.FreqDist(all_tokens)
most_frequent_words = [x[0] for x in f_distribution.most_common(N)]

result_bigrams = {}

bigram_measures = nltk.collocations.BigramAssocMeasures()
for key, value in result_tokens.items():
    text = result_tokens[key]
    finder = BigramCollocationFinder.from_words(text,window_size = win_size)    
    finder.apply_word_filter(lambda w: w not in most_frequent_words)  #solo busco los bigramas entre las palabras mas frecuentes
    items = list(finder.ngram_fd.items())
    
    for item in items:
        if item[0][0] != item[0][1]: #solo agrego los bigramas de palabras diferentes
        
            if item[0] in result_bigrams:
                result_bigrams[item[0]] = result_bigrams[item[0]] + item[1]
            else:
                result_bigrams[item[0]] = item[1]
                
#conservo solamente los bigramas cuya frecuencia es o está por encima de 6 (para valores menores la medida de asociacion es inestable)
thresholded_bigram = {k: v for k, v in result_bigrams.items() if v >= threshold}

#calculo la medida de asociacion pointwise mutual information
#preparo un diccionario con las frecuencias de cada palabra de interés en el corpus total
word_frequency = {}
for w in most_frequent_words:
    word_frequency[w] = all_tokens.count(w)

#pmi calculation
pmi_bigram = {}
for bk,bv in thresholded_bigram.items():
    fxy = bv/(win_size - 1)
    pxy = fxy/vocabulary_n
    
    px = word_frequency[item[0][0]]/vocabulary_n
    py = word_frequency[item[0][1]]/vocabulary_n
        
    pmi = pxy/(px*py)
    
    pmi_bigram[bk] = pmi
    
highest_associated_pair = max(pmi_bigram, key = pmi_bigram.get)

print('El par más asociado es ' + str(highest_associated_pair) + ' cuyo pmi es ' + str(pmi_bigram[highest_associated_pair]))
```

### 1. Procesar el texto, tokenizando eliminando signos de puntuación.


```python
import urllib3
from bs4 import BeautifulSoup
from nltk.tokenize import RegexpTokenizer

url = "http://www.gutenberg.org/cache/epub/1228/pg1228.txt"
http = urllib3.PoolManager()
response = http.request('GET', url)
raw=response.data
tokenizer = RegexpTokenizer(r'\w+')
tok_species=tokenizer.tokenize(raw)
```

### 2. Siguiendo el artículo de la sección, calcular la autocorrelación para estimar la distribución de la palabra a lo largo del texto.



```python
from scipy.signal import gaussian
import matplotlib.pyplot as plt
from scipy.ndimage.interpolation import shift

def autocorr(x, t=1):
    return np.corrcoef(np.array([x[0:len(x)-t], x[t:len(x)]]))

word=('for','plant','instinct')

for word_i in word:

    rate=np.zeros(len(tok_species))
    
    indices = [i for i, x in enumerate(tok_species) if x == word_i]

    for index in indices:
    
        rate += shift(gaussian(len(tok_species), std=50), index-len(tok_species)/2, cval=0)

    a_corr=np.zeros(140000)

    for tau in range(140000):
        a_corr[tau]=autocorr(rate,tau)[0][1]

    plt.plot(rate)    
    plt.title(word_i)
    plt.ylabel("Rate")
    plt.xlabel("Position")    
    plt.show()
    plt.plot(a_corr)
    plt.title(word_i)
    plt.ylabel("Autocorrelation")
    plt.xlabel("tau")
    plt.show()
```

Conclusión: Se observa que las palabras menos frecuentes se concentran en regiones específicas del texto, en donde además se concentran por lo que su autocorrelación decae más lentamente. La caída rápida de la autocorrelación indica una menor informaciòn aportada por la precencia de la palabra en un momento concreto (a diferencia de un randomizado del texto). De aquí se puede inferir que las palabras que se concentran en partes específicas del texto aportan información crucial para estas secciones.

### 3. Calcular la entropía de una selección de 100 palabras que abarquen el rango de frecuencia de aparición en el libro, es decir elegir palabras que son muy frecuentes y otras de baja frecuencia.

```python
from collections import Counter

total_len=len(tok_species)
part_len=len(tok_species)/64

most_common=Counter(tok_species).most_common(len(set(tok_species)))
words=np.zeros(100)
words_temp=most_common[30:100]+most_common[int(len(most_common)/25):int(len(most_common)/25)+10]+most_common[int(len(most_common)/15):int(len(most_common)/15)+10]+most_common[len(most_common)/4:len(most_common)/4+10]
words = [x[0] for x in words_temp]

entropy_by_word=np.zeros(len(words))
freq_by_word=np.zeros(len(words))

for j, word_i in enumerate(words):
        
    indices = [i for i, x in enumerate(tok_species) if x == word_i]  # frecuencia de aparición en el texto tota

    entropy=0

    for i in range(64):
        
        part_n=tok_species[i*part_len:i*part_len+(part_len-1)]
        indices_n = [i for i, x in enumerate(part_n) if x == word_i]  # frecuencia de aparición en el texto n
        if len(indices_n)>0:
            entropy+=(float(len(indices_n))/float(len(indices)))*np.log2(float(len(indices_n))/float(len(indices)))
    
    entropy_by_word[j]=entropy
    freq_by_word[j]=len(indices)
```

```python
plt.title('Entropia de palabras mas y menos frecuentes')
plt.ylabel("Entropy")
plt.xlabel("Freq")   
plt.plot(freq_by_word,-entropy_by_word,"o")
plt.show()
```

### 4. Calcular la entropía de la distribución de palabras seleccionadas anteriormente, randomizando su posición en el texto. Comparar con los resultados del punto anterior.


```python
from random import shuffle

shuffle(tok_species)

entropy_by_word=np.zeros(len(words))
freq_by_word=np.zeros(len(words))

for j, word_i in enumerate(words):
        
    indices = [i for i, x in enumerate(tok_species) if x == word_i]  # frecuencia de aparición en el texto tota

    entropy=0

    for i in range(64):
        
        part_n=tok_species[i*part_len:i*part_len+(part_len-1)]
        indices_n = [i for i, x in enumerate(part_n) if x == word_i]  # frecuencia de aparición en el texto n
        if len(indices_n)>0:
            entropy+=(float(len(indices_n))/float(len(indices)))*np.log2(float(len(indices_n))/float(len(indices)))
    
    entropy_by_word[j]=entropy
    freq_by_word[j]=len(indices)
            
```

```python
plt.title('Entropia de palabras randomizadas')
plt.ylabel("Entropy")
plt.xlabel("Freq")   
plt.plot(freq_by_word,-entropy_by_word,"o")
plt.show()
```

Conclusión: En la relación entre entrpía y frecuencia se observa que mientras menos frecuentes son las palabras, su valor de entropía más se aleja del esperado para un texto con palabras randomizadas. Esto indica que las palabras menos frecuentes tienen menos entropía que la aleatoria y por tanto aportan más información respecto del texto.

```python
#cargo wordsim353 en un pandas data frame
#file_name = 'H:\\Cursos Doctorado\\CienciaDeDatos\\6.NLP\\wordsim353\\' + 'combined.csv'
file_name = 'D:\\Euge\\Cursos\\CienciaDeDatos\\TP6\\combined.csv'
wordsim353_df = pd.read_csv(file_name)
wordsim353_df.columns = ['word1', 'word2','simscore']

row_nr = wordsim353_df.shape[0]

vector_dimension = 300
```

```python
#PRETRAINED LSA WITH WIKIPEDIA
#MODELO ENTRENADO CON WIKIPEDIA OBTENIDO DE http://deeptutor2.memphis.edu/Semilar-Web/public/lsa-models-lrec2014.html
path_to_pretrained_lsa = 'H:\\Cursos Doctorado\\CienciaDeDatos\\6.NLP\\Wiki 6\\Wiki 6\\'

lsa_file = path_to_pretrained_lsa + 'lsaModel'
voc_file = path_to_pretrained_lsa + 'voc'

mat = np.loadtxt(lsa_file,delimiter=' ')

with open(voc_file) as f:
    voc = f.read().split()   

dictionary = {}
for r in range(0,mat.shape[0]):
    dictionary[voc[r]] = mat[r,:]
```

```python
equivalence = WordNetLemmatizer()    
lsa_sim = np.zeros((row_nr,1))

for r in range(0,row_nr):    
    word1 = wordsim353_df.loc[r,'word1']
    word2 = wordsim353_df.loc[r,'word2']
    
    exc = 0
    
    if word1 in dictionary.keys():
        w1 = dictionary[word1] #obtengo el id correspondiente a la palabra
    else:
        nword1 = equivalence.lemmatize(word1) 
        if nword1 in dictionary.keys():
            w1 = dictionary[nword1] #obtengo el id correspondiente a la palabra
        else:
            exc = 1
    
    
    if exc == 0:
        if word2 in dictionary.keys():
            w2 = dictionary[word2] #obtengo el id correspondiente a la palabra
        else:
            nword2 = equivalence.lemmatize(word2) 
            if nword2 in dictionary.keys():
                w2 = dictionary[nword2] #obtengo el id correspondiente a la palabra
            else:
                exc = 1           
    
    if exc == 1:
        lsa_sim[r] = np.nan
    else:
        lsa_sim[r] = np.dot(matutils.unitvec(w1), matutils.unitvec(w2))    
```

```python
#agrego columnas al wordsim353_df data frame     
wordsim353_df['lsa'] = lsa_sim
notnans = pd.notnull(wordsim353_df['lsa'])
[rlsa,pvallsa] = stats.pearsonr(wordsim353_df['simscore'][notnans],wordsim353_df['lsa'][notnans])      
ax = plt.gca()
sns.regplot(x="simscore", y="lsa", data=wordsim353_df);
ax.set_title("LSA: r = " + str(rlsa) + ", p = " + str(pvallsa))
plt.show() #no se si va
```

```python
equivalence = WordNetLemmatizer()    
w2v_sim = np.zeros((row_nr,1))

for r in range(0,row_nr):    
    word1 = wordsim353_df.loc[r,'word1']
    word2 = wordsim353_df.loc[r,'word2']
    
    exc = 0
    #word2vec
    if word1 in model.vocab:
        w1 = model.word_vec(word1) #obtengo el id correspondiente a la palabra
    else:
        nword1 = equivalence.lemmatize(word1) 
        if nword1 in model.vocab:
            w1 = model.word_vec(nword1) #obtengo el id correspondiente a la palabra
        else:
            exc = 1
        
    if exc == 0:
        if word2 in model.vocab:
            w2 = model.word_vec(word2) #obtengo el id correspondiente a la palabra
        else:
            nword2 = equivalence.lemmatize(word2) 
            if nword2 in model.vocab:
                w2 = model.word_vec(nword2) #obtengo el id correspondiente a la palabra
            else:
                exc = 1           
    
    if exc == 1:
        w2v_sim[r] = np.nan
    else:
        w2v_sim[r] = np.dot(matutils.unitvec(w1), matutils.unitvec(w2))   
```
